```python
import sympy as sp
sp.init_printing(use_unicode=True)
```

```python
x = sp.Symbol('x')
f = sp.sin(x) * x
```

```python
# 求极限
limit = sp.limit(f, x, float('inf'))
print(limit)
```

```python
# 求导数
derivative = sp.diff(f,x)
print(derivative)
```

```python
# 求定积分
definite_integral = sp.integrate(f, (x,0,sp.pi))
print(definite_integral)
```

```python
# 求不定积分
integral = sp.integrate(f,x)
print(integral)
```

```python
# 级数展开
series_res = f.series(x, 0, 6)
print(series_res)
```

```python
# section 7.6, 92
f = 1 / ((sp.sin(x)+sp.cos(x))**2)
integrate_res_1 = sp.integrate(f, x)
print(integrate_res_1)
```

```python
# section 7.7, Example 1
f = 2 * sp.pi * x * sp.atan(x)
def_integrate_res_1 = sp.integrate(f, (x, 0, 1))
print(def_integrate_res_1)
```

```python
# section 7.7, Example 5
f = x * sp.sqrt(x**2 + 2*x + 4)
integrate_res_2 = sp.integrate(f, x)
print(integrate_res_2)
```

```python
# section 7.7, Example 6
f = x * (x**2 + 5)**8
integrate_res_3 = sp.integrate(f, x)
print(integrate_res_3)
```

```python
# 经典永流传
ans = [sp.integrate(1 / (x**i + 1), x) for i in range(1, 8)]
for i in ans:
    print(i)
```
