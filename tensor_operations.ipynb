```python
import torch
```

```python
a = torch.tensor([[1,2],[3,4]])
b = torch.tensor([[5,6],[7,8]])
```

```python
# 加法
print(a + b)
print(torch.add(a, b))
print(a.add(b))
# a.add_(b) 原地操作
```

```python
# 减法
print(a-b)
print(torch.sub(a,b))
print(a.sub(b))
# a.sub_(b)
```

```python
# 乘法（对应元素直接）
print(a*b)
print(torch.mul(a,b))
print(a.mul(b))
# a.mul_(b)
```

```python
# 除法（对应元素相除）
print(a / b)
print(torch.div(a,b))
print(a.div(b))
# a.div_(b)
```

```python
# 矩阵乘法
# 二阶矩阵
print(a@b)
print(torch.mm(a,b))
print(a.mm(b))
print(torch.matmul(a,b))
print(a.matmul(b))

# 高阶矩阵（只作用于最后两个维度）
c = torch.ones(2,3,4)
d = torch.ones(4,3)
print(torch.matmul(c,d))
```

```python
# 幂运算
print(a**2)
print(torch.pow(a,2))
print(a.pow(2))
print()
# a.pow_(2)
```

```python
# 开方运算
print(a.sqrt())
# a.sqrt_()
```

```python
# 指数运算
print(torch.exp(a))
print(a.exp())
# a.exp_()

# 对数运算
print(torch.log(a)) # 底数为e
```

```python
x = torch.rand(2,3) + 3
print(x)
```

```python
# 取整运算
print(torch.floor(x)) # 向下取整
print(x.ceil()) # 向上取整
print(x.round()) # 四舍五入
print(torch.trunc(x)) # 取整数部分
print(torch.frac(x)) # 取小数部分
```

```python
print(a+b)
print(a+c)
print(b+c)
```

比较运算

```python
a = torch.tensor([[1,2],[3,4]])
b = torch.tensor([[1,2],[4,3]])
print(torch.eq(a, b)) # 按照元素分别比较,形状不同就会报错
print(torch.equal(a, b)) # 整体比较,只有它是整体比较
print(torch.gt(a, b)) # greater than
print(torch.ge(a, b)) # greater and equal
print(torch.lt(a, b)) # less than
print(torch.le(a, b)) # less and equal
print(torch.ne(a, b)) # not equal
```

排序

```python
c = torch.randint(0,100,(2,3,2))
print(c)
print(torch.sort(c, dim=2).values)
# print(torch.sort(c, descending=True).values) # 降序
```

```python
# topk
d = torch.randint(0,10,(2,5))
print(d)
print(torch.topk(d, k=2, dim=0))
print(torch.kthvalue(d, k=2, dim=1)) # 第k小的数字
print(torch.kthvalue(d, k=2, dim=0))
```
