```python
import torch
a = torch.rand(2,2,2)
```

```python
# 平均值
print(a)
print(torch.mean(a, dim=2))
print(torch.sum(a, dim=1))
print(torch.prod(a, dim=0))
```

```python
# 最大/小值的索引位置
print(torch.argmax(a, dim=0))
print(torch.argmin(a, dim=0))
```

```python
# 其他
print(torch.std(a))
print(torch.var(a))
print(torch.median(a))
print(torch.mode(a))

```

```python
a = torch.rand(2,2,2) * 10
print(torch.histc(a, 6, 0, 0))
```
